## Configuración
Colocar el archivo SQL en `../data/` si abre desde `notebooks/`.

```python
import sys
sys.path.append('..')
from etl import ETLProductos, DBConnection
print('Módulos cargados')
```

```python
file_path = '../data/productos.sql'
columns = ['nombre', 'codigo', 'proveedor', 'categoria', 'empresa', 'stock', 'precio', 'costo', 'otro']
output_name = 'productos_limpio'
db = DBConnection(host='localhost', user='root', password='', database='negocio')
try:
    db.connect()
except Exception:
    db = None
```

```python
etl = ETLProductos(file_path=file_path, columns=columns, output_name=output_name, db_conn=(db.conn if db else None))
etl.extract()
etl.transform()
etl.save_csv()
etl.save_sqlite()
print('ETL completado para', output_name)
```
