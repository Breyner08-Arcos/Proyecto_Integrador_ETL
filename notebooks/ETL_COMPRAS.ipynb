## Configuración
Colocar el archivo SQL en `../data/` si abre desde `notebooks/`.

```python
import sys
sys.path.append('..')
from etl import ETLCompras, DBConnection
print('Módulos cargados')
```

```python
file_path = '../data/compras.sql'
columns = ['fecha', 'codigo', 'producto', 'cantidad', 'precio_unitario']
output_name = 'compras_limpio'
db = DBConnection(host='localhost', user='root', password='', database='negocio')
try:
    db.connect()
except Exception:
    db = None
```

```python
etl = ETLCompras(file_path=file_path, columns=columns, output_name=output_name, db_conn=(db.conn if db else None))
etl.extract()
etl.transform()
etl.save_csv()
etl.save_sqlite()
print('ETL completado para', output_name)
```
